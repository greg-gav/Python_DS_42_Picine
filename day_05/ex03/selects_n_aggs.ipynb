```python
import pandas as pd
```

## load JSON

```python
df = pd.read_json(
    "../data/auto.json",
)
df = df.set_index("CarNumber", copy=False)
```

## make selects

```python
df.loc[df['Fines'] > 2100]
```

```python
df.loc[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

```python
df.loc[df['Model'].isin(['Focus','Corolla'])]
```

```python
df.iloc[df.index.isin(['Y7689C197RUS','92928M178RUS','7788KT197RUS','H115YO163RUS','X758HY197RUS'])]
```

## aggregations with the make and the model


```python
df.groupby('Make').agg({'Fines' : 'median'})
```

```python
df.groupby(['Make', 'Model']).agg({'Fines' : 'median'})
```

```python
df.groupby(['Make', 'Model']).agg({'Fines' : 'count'})
```

```python
pd.set_option("display.precision", 1)
df.groupby(['Make', 'Model']).agg({'Fines' : ['min', 'max']})
```

```python
df.groupby(['Make', 'Model']).agg({'Fines' : 'std'})
```

## aggregations with car number

```python
nums = df.groupby(['CarNumber']).agg({'Fines' : 'count'})
nums.sort_values('Fines', ascending=False)
```

```python
df.iloc[df.index == "Y7689C197RUS"]
```

```python
nums = df.groupby(['CarNumber']).agg({'Fines' : 'sum'})
nums.sort_values('Fines', ascending=False)
```

```python
df.iloc[df.index == "X758HY197RUS"]
```
